```python
import fitz  # PyMuPDF
import json
import re

# Load the PDF
pdf_path = "Lugat_i_Seri_Seriyye_Sicilleri_Sozlugu.pdf"
document = fitz.open(pdf_path)

# Regex pattern to detect entries
# This pattern is adjusted to capture the term before Arabic parenthesis and the definition after the colon
entry_pattern = re.compile(r"([^\(\)]+) \([\u0600-\u06FF]+\)\s*:\s*(.*)")

# Extract text from each page and gather the dictionary entries
dictionary_entries = {}

# Process each page
for page_num in range(document.page_count):
    page = document.load_page(page_num)
    text = page.get_text("text")

    # Process text to extract dictionary entries
    lines = text.split("\n")

    for line in lines:
        line = line.strip()
        match = entry_pattern.match(line)
        if match:
            # Extract the term and definition
            term = match.group(1).strip()
            definition = match.group(2).strip()
            dictionary_entries[term] = definition
            print(f"Match found: {term}: {definition}")  # Print each matched entry

# Create a JSON object
dictionary_json = json.dumps(dictionary_entries, ensure_ascii=False, indent=2)

# Save the JSON data to a file
output_path = "ottoman_court_records_dictionary.json"
with open(output_path, "w", encoding="utf-8") as file:
    file.write(dictionary_json)

print("JSON dictionary has been saved:", output_path)
```

```python
import json
import re

input_file_path = "ottoman_court_records_dictionary.json"
output_file_path = "cleaned_ottoman_court_records_dictionary.json"
# Read the content of the JSON file
with open(input_file_path, "r", encoding="utf-8") as file:
    data = json.load(file)

# A list to store all dictionary entries
combined_entries = {}
    
# Regular expression to find JSON strings
json_pattern = re.compile(r'\{.*?\}', re.DOTALL)

# Process each item in the loaded data
for item in data:
    # Search for JSON formatted text
    matches = json_pattern.findall(item)
    for match in matches:
        try:
            # Convert the JSON string to a dictionary
            entry_dict = json.loads(match)
            # Combine all entries into one dictionary
            combined_entries.update(entry_dict)
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON: {e}")
with open(output_file_path, "w", encoding="utf-8") as file:
    json.dump([combined_entries], file, ensure_ascii=False, indent=2)

print(f"Cleaned JSON has been saved to '{output_file_path}'")
```

```python
import json

def count_tokens_in_json(file_path):
    # Load the JSON data from the file
    with open(file_path, "r", encoding="utf-8") as file:
        data = json.load(file)
    
    # Initialize token count
    total_tokens = 0

    # Iterate through each dictionary in the list
    for dictionary in data:
        # Iterate over each key and definition in the dictionary
        for key, definition in dictionary.items():
            # Tokenize the key and definition by splitting on whitespace
            key_tokens = key.split()
            definition_tokens = definition.split()

            # Add the number of tokens from key and definition to the total count
            total_tokens += len(key_tokens) + len(definition_tokens)
    
    return total_tokens

# Example usage
file_path = "cleaned_ottoman_court_records_dictionary.json"
total_tokens = count_tokens_in_json(file_path)
print(f"Total number of tokens in the file: {total_tokens}")
```

```python
import json
from transformers import GPT2Tokenizer

def calculate_openai_tokens(file_path):
    # Load the JSON data from the file
    with open(file_path, "r", encoding="utf-8") as file:
        data = json.load(file)
        
    # Initialize the tokenizer
    tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
    
    total_tokens = 0

    # Iterate through each dictionary in the list
    for dictionary in data:
        # Iterate over each key and definition in the dictionary
        for key, definition in dictionary.items():
            # Count tokens for both the key and the definition
            key_tokens = tokenizer.encode(key, add_special_tokens=False)
            definition_tokens = tokenizer.encode(definition, add_special_tokens=False)

            # Add the number of tokens from key and definition to the total count
            total_tokens += len(key_tokens) + len(definition_tokens)

    return total_tokens


# Load the dictionary
with open('cleaned_ottoman_court_records_dictionary.json', 'r', encoding='utf-8') as file:
    dictionary = json.load(file)

# Convert dictionary to a string to include in the prompt
dictionary_str = json.dumps(dictionary[0], ensure_ascii=False, indent=2)


# Example usage
file_path = "cleaned_ottoman_court_records_dictionary.json"
total_tokens = calculate_openai_tokens(file_path)
print(f"Total number of OpenAI API tokens in the file: {total_tokens}")
```
