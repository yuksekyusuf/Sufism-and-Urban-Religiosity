```python
df['year'] = df['date_miladi'].str.extract(r'M\. (\d{4})').astype(float)

# NaN değerlerini kaldırın
df = df.dropna(subset=['year'])

# Yıllık bazda dava sayısını hesaplayın
yearly_case_counts = df.groupby(['year', 'court_title']).size().unstack(fill_value=0)

# Yıllık dava dağılımını gösteren bir grafik oluşturun
plt.figure(figsize=(15, 8))
yearly_case_counts.plot(kind='line', marker='o')
plt.title('Yıllık Dava Dağılımı Mahkeme İsimleri ile')
plt.xlabel('Yıl')
plt.ylabel('Dava Sayısı')
plt.grid(True)
plt.legend(title='Mahkemeler', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
