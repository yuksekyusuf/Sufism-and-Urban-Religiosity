```python
import urllib.request, urllib.error, urllib.parse, time
import os.path
```

```python
def downloadPages(n, number):
    url = "http://www.kadisicilleri.org/arascl/ayrmetin.php?idno="
    cwd = os.getcwd()
    
    for item in range(n, number):
        new_url = url+"{}".format(item)
        response = urllib.request.urlopen(new_url)
        webContent = response.read()
        
        save_path = cwd+'/urls'
        filename = "{}".format(item)
        completeName = os.path.join(save_path, filename)
        print()
        
        f = open(completeName + ".html", 'wb')
        f.write(webContent)
        f.close
        print(item)
        time.sleep(0.5)
```

```python
downloadPages(53440, 54671)
```
