````python
def automate_ner(case_input, court_title, case_id):
    prompt_messages = [
        {"role": "system", "content": prompt},
        {"role": "user", "content": case_input}
    ]
    try:
        completion = client.chat.completions.create(
            model="gpt-4o",
            messages=prompt_messages,
            temperature=0.05,
            n=1,
            stop=None
        )
        output_str = completion.choices[0].message.content
        cleaned_output_str = output_str.replace("```python\n", "").replace("\n```", "")
        output_dict = eval(cleaned_output_str)
        
        # Add unique person_id to each person
        for idx, person in enumerate(output_dict.get('persons', []), start=1):
            person_id = f"{court_title}_{case_id}_{idx}"
            person['person_id'] = person_id
        
        return output_dict  # Return the structured NER result
    except Exception as e:
        print(f"An error occurred: {e}")
        return {'error': str(e)}
````

````python
def process_dataframe_ner(df, summary_column_name, court_title_column_name, case_id_column_name):
    ner_results = []

    for index, row in df.iterrows():
        case_input = row[summary_column_name]
        court_title = row[court_title_column_name]
        case_id = row[case_id_column_name]  # Capture the unique ID for each case
        ner_result = automate_ner(case_input, court_title, case_id)  # Process each case entry
        
        if 'error' not in ner_result:
            for person in ner_result.get('persons', []):
                person['legal_case_type'] = ner_result.get('legal_case_type', 'N/A')
                person['row_index'] = index  # Track original row index
                person['case_unique_id'] = case_id  # Attach the unique ID to each person's entry
                ner_results.append(person)
        else:
            print(f"Error processing row {index}: {ner_result['error']}")
    
    # Convert the collected results into a DataFrame
    ner_df = pd.DataFrame(ner_results)
    return ner_df
````
