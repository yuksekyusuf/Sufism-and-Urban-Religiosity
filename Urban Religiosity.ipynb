````python
def automate_ner(case_input):
    prompt_message = [
        {"role": "system", "content": prompt},
        {"role": "user", "content": case_input}
    ]
    
    try:
        completion = client.chat.completions.create(
            model="gpt-4-turbo-preview",
            messages=prompt_messages,
            temperature=0.05,
            n=1,
            stop=None
        )
        output_str = completion.choices[0].message.content
        cleaned_output_str = output_str.replace("```python\n", "").replace("\n```", "")
        output_dict = eval(cleaned_output_str)
        return output_dict
    except Exception as e:
        print(f"An error occurred: {e}")
        return {'error': str(e)}
````

````python
def process_dataframe_ner(df, text_column, column_id):
    ner_results = []
    for index, row in df.iterrows():
        case_input = row[text_column]
        case_id = row[column_id]
        ner_result = automate_ner(case_input)
        print(ner_result)
        if 'error' not in ner_result:
            for person in ner_result.get('persons', []):
                person['legal_case_type'] = ner_result.get('legal_case_type', 'N/A')
                person['row_index'] = index
                person['case_unique_id'] = case_id
                ner_results.append(person)
        else:
            print(f"Error processing row {index}: {ner_result['error']}")
    ner_df = pd.DataFrame(ner_results)
    return ner_df
````

````python
def perform_ner(text):
    prompt_messages = [
        {"role": "system", "content": prompt},
        {"role": "user", "content": text}
    ]
    try:
        completion = client.chat.completions.create(
            model="gpt-4-turbo-preview",
            messages=prompt_messages,
            temperature=0.05,
            n=1,
            stop=None
        )
        ner_result = eval(completion.choices[0].message.content)
        return ner_result
    except Exception as e:
        print(f"An error occurred while processing row {index}: {e}")
        return (index, {'error': str(e)})
````

````python
ner_results = []

for index, row in df.iloc[14:18].iterrows():
    ner_result = perform_ner(row['case_text_summary'])

    # Check for errors in ner_result
    if 'error' not in ner_result:
        for person in ner_result.get('persons', []):
            # Supplement each person's dictionary with additional data
            person['legal_case_type'] = ner_result.get('legal_case_type', 'N/A')
            person['row_index'] = index
            ner_results.append(person)
    else:
        # Log the error and possibly handle or record the index for reprocessing
        print(f"Error processing row {index}: {ner_result['error']}")

ner_df = pd.DataFrame(ner_results)
````
